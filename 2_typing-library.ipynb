```python
from pydantic import BaseModel
```

```python
from typing import List, Dict, Optional
```

```python
class Patient(BaseModel) :
    name : str
    age : int
    wieght : float
    married : Optional[bool] = False
    allergies : List[str]
    contact: Dict[str,str]
```

```python
p1 = Patient(**{
    'name':'shri',
    'age':22,
    'wieght':50.5,
    # 'married':False,
    'allergies':['dust','aaloo'],
    'contact' : {'mobile':'+91-97555-62462'}
})
```

```python
def insert(p : Patient) :
    print(p.allergies)
    print(p.contact['mobile'])
    print(p.married)

insert(p1)
```
