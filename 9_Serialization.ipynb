```python
from pydantic import BaseModel
from typing import List, Dict, Optional
```

```python
class Patient(BaseModel) :
    name : str
    age : int
    wieght : float
    married : Optional[bool] = False
    allergies : List[str]
    contact: Dict[str,str]
```

```python
p1 = Patient (
    name = "Shri",
    age = 30,
    wieght = 70.5,
    married = True,
    allergies = ["Peanuts", "Shellfish"],
    contact = {"email": "shri@example.com"}
)
```

```python
p1.model_dump()
p1.model_dump_json()
```
