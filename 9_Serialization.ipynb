```python
from pydantic import BaseModel
from typing import List, Dict, Optional
```

```python
class Patient(BaseModel) :
    name : str
    age : int
    wieght : float
    married : Optional[bool] = False
    allergies : List[str]
    contact: Dict[str,str]
```

```python
p1 = Patient (
    name = "Shri",
    age = 30,
    wieght = 70.5,
    # married = True,
    allergies = ["Peanuts", "Shellfish"],
    contact = {"email": "shri@example.com"}
)
```

```python
py_dict = p1.model_dump()
py_dict
```

```python
json_obj = p1.model_dump_json()

type(py_dict)
```

```python
type(json_obj)
```

```python
# along with include and exclude parameters, we can also use by_alias to control the keys in the output

p1.model_dump(include={"name", "age"}, by_alias=True)
```

```python
p1.model_dump(exclude={"wieght", "married"}, by_alias=True)
```

```python
p1.model_dump(exclude_unset=True)

# as you can see the married option is excluded
```
