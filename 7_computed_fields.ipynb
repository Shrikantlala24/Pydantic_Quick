```python
from pydantic import BaseModel, computed_field
from typing import List, Optional
```

```python
class User(BaseModel) :
    name: str
    age: int
    height: float
    weight: float


    @computed_field
    @property
    def bmi(self) -> float:
        return self.weight / (self.height ** 2)
```

```python
u1 = User(
    name = 'Shri',
    age = 30,
    height = 1.75,
    weight = 70
)
print(u1.bmi)
```
