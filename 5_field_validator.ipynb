```python
from pydantic import BaseModel, field_validator, EmailStr, AnyUrl
from typing import (
    Optional, 
    List, 
    Dict,
    Annotated
    )
```

# Field Validator
- we'll create our own datatypes by modifying the fields available in class

```python
class User(BaseModel):
    name : str
    age : int
    email : str
    website : Optional[AnyUrl] = None

    # ---------------------------------------------------------------------------------------------  Block 1
    
    @field_validator('email')
    @classmethod
    def hutt(cls,value):
        valid_domains = ['hdfc.com','icici.com']
    
        if (value.split('@')[-1]) in valid_domains:
            return value
            
        raise ValueError(f'invalid email - enter the from {valid_domains}')

    # --------------------------------------------------------------------------------------------- Block 2

    @field_validator('name')
    @classmethod
    def Upper_case(cls,value):
        return value.upper()

    # --------------------------------------------------------------------------------------------- Block 3

    @field_validator('age', mode='before')
    @classmethod
    def func(cls,value):
        if 0 < value < 100:
            return value
        raise ValueError('Invalid age')
```

```python
u1 = User(**{
    # it will be capitalized
    'name': 'shri', 
    
    # it will type checked, before type-coercion
    'age': 30, 
    
    # valid email check
    'email': 'shri@hdfc.com', 
    
    'website': 'https://www.example.com'
    })
```
