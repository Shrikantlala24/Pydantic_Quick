```python
from pydantic import BaseModel
from typing import List, Optional
```

```python
class Address(BaseModel) :
    city : str
    state : str
    country : str
```

```python
class User(BaseModel):
    name : str
    age : int
    address : Address
```

```python
u1 = User(
    name="Alice",
    age=30,
    address=Address(
        city="New York",
        state="NY",
        country="USA"
    )
)
```

```python
print(u1.address.city)
print(u1.address)
```
