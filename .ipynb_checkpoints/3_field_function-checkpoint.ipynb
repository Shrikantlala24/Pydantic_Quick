```python
from pydantic import BaseModel, Field
from typing import List, Optional, Annotated, Any
```

```python
class Patient(BaseModel):
    name: str
    age: int = Field(gt=0)
    allergies: Optional[List[str]] = Field(
        default=None,
        max_length=5,
    )
```

```python
# help(Field)
```

```python
def insert(p : Patient) :
    print(p.name)
    print(p.age)
    print(p.allergies)
```

```python
p1 = Patient(**{
    'name':'Shri',
    'age':30,
    'allergies':['dust','smoke','peanuts']
})
```

```python
insert(p1)
```
