```python
from pydantic import BaseModel,model_validator
from typing import Dict
```

```python
class Patient(BaseModel) :
    name: str
    age : int
    contact : Dict[str,str]

    # main block
    @model_validator(mode='after')
    def emergency_contact(cls,model):
        if (model.age > 60) and ('emergency' not in model.contact ) :
            raise ValueError('\n\n emergency contact has not entered\n\n')

        return model
```

```python
p1= Patient(**{
    'name':'shri',
    'age':'30',
    'contact': {
        'mob':'234456',
    }
})
```

```python
p2 = Patient(**{
    'name':'nanu',
    'age':'65',
    'contact': {
        'mob':'234456',
        'emergency':'13214'
    }
})
```

```python
p3 = Patient(**{
    'name':'dadu',
    'age':'65',
    'contact': {
        'mob':'234456',
        'emergency':'13214'
    }
})
```
