## Practice Problem: Book Catalog API Schema

You’re designing a **Book Catalog API**. Create a Pydantic model for the endpoint:

`POST /books/`

This endpoint should accept a JSON body describing a book.

### Requirements

Define a `Book` model with these fields:

- `title`:  
  - `str`  
  - required  
  - must be at least 1 character and at most 100 characters  

- `author`:  
  - `str`  
  - required  
  - at most 60 characters  

- `published_year`:  
  - `int`  
  - required  
  - must be greater than `0` and less than or equal to the current year (for now, just use `le=2026` as a simple constraint)  

- `price`:  
  - `float`  
  - required  
  - must be strictly greater than `0`  

- `genres`:  
  - `List[str]`  
  - optional  
  - if provided, limit to maximum 5 items using `max_length` on the list  

- `in_stock`:  
  - `bool`  
  - required  
  - default to `True`  

- `rating`:  
  - `float`  
  - optional  
  - if provided, must be between 0 and 5 (inclusive)  

Use only things you already know:

- `BaseModel`  
- standard types (`str`, `int`, `float`, `bool`, `List[str]`)  
- `Optional[...]`  
- `Field(...)` with constraints like `gt`, `ge`, `le`, `min_length`, `max_length`, `default`, `default_factory`  

You do **not** need custom validators yet.

***


```python
from pydantic import BaseModel, Field
from typing import List, Dict, Annotated, Optional
```

```python
class Book(BaseModel) :
    title :Annotated[
        str, 
        Field(...,min_length=1,max_length=100)
        ]
    author : Annotated[
        str, 
        Field(...,max_length=60)
        ]
    published_year : Annotated[
        int,
        Field(...,gt=0,le=2026)
        ]
    price :Annotated[
        float,
        Field(...,gt=0,strict=True)
        ]
    genres : Annotated[
        Optional[List[str]],
        Field(default=None, max_length=5)
        ]
    in_stock : bool = True
    
    rating : Annotated[
        Optional[float], Field(ge=0,le=5)
        ]
```

```python
b1 = Book(**{
    'title':'Atomic Habits',
    'author':'James Clear',
    'published_year':'2026',
    # price should be strictly int, pydantic won't implicitly convert it into 'int' {if strict=True}
    'price':499.5,
    'genres':['personality dev', 'habits'],
    'in_stock': False,
    'rating':4.9
})
```
