```python
from pydantic import BaseModel
```

```python
class Patient(BaseModel) :
    name : str
    age : int
```

```python
def insert(p : Patient):
    print(p.name)
    print(p.age)
    print("Inserted")
```

```python
p_manual = {'name': 'Shrikant', 'age':'21'}
p1 = Patient(**p_manual)
```

```python
insert(p1)
```
